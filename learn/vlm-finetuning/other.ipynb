```python
import boto3
from time import sleep

bedrock = boto3.client("bedrock")

# Wait until job complete.
# You can also view the console for status: https://console.aws.amazon.com/bedrock/home#/import-models
status = "InProgress"
while status != "Completed":
    response = bedrock.get_model_import_job(jobIdentifier="arn:aws:bedrock:us-east-1:023979239971:model-import-job/uth83abz9uzg")
    print(response)
    sleep(5)
    status = response["status"]
```
